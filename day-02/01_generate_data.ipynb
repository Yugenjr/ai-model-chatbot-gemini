### Generate synthetic "messy" e-commerce data

```python
data = {
    "transaction_id": range(1, 101),
    "customer_age": np.random.choice([np.nan, 25, 34, 17, 85, 120], 100), # Includes NaNs, underage, and unrealistic ages
    "purchase_amount": np.random.normal(50, 40, 100), # Will generate some negative amounts
    "status": np.random.choice(["completed", "pending", "failed", "UNKNOWN"], 100)
}
```

```python
print(data)
```

```python
df = pd.DataFrame(data)
print(df.head())
```

```python
from pathlib import Path
from IPython.display import FileLink

output_path = Path.cwd() / "messy_ecommerce_data.csv"
df.to_csv(output_path, index=False)
print(f"messy_ecommerce_data.csv created successfully at: {output_path.resolve()}")
print(f"File exists: {output_path.exists()}")
FileLink(str(output_path))
```
